```python
df_product = df[df.Category=='PRODUCTIVITY']
df_product = df_product.drop(['Genres','Size', 'Price', 'Last Updated', 'Current Ver', 'Android Ver', 'Content Rating'], axis=1)

df_product.head()
```

Alguns aplicativos de produtividade conhecidos, como **Microsoft Word** (editor de textos), **Adobe Acrobat Reader** (Leitor de PDF), **Google Drive** (Armazenamento na nuvem)

Organiza e limpa o DataFrame df_educ de aplicativos educacionais da seguinte forma:

*   Ordena os aplicativos pela quantidade de avaliações (em ordem decrescente).

*   Remove aplicativos duplicados, garantindo que cada um apareça apenas uma vez.

Ordenação dos dados pela quantidade de avaliações (de maior para menor), remove valores nulos e, em seguida, filtra os produtos que receberam uma avaliação de 5 estrelas.

```python
df_product.sort_values('Reviews', ascending=False)
df_product.dropna(inplace=True)
df_product.loc[df_product['Reviews']=='5']
```

Concatenação dos dados, formando um só, assim criamos o dataframe que iremos analisar a partir de agora

```python
df_educ_product = pd.concat([df_educ, df_product])
df_educ_product
```

Descrições estatísticas do nosso dataframe para entender como estão distribuídas suas principais características estatísticas, como: quantidade, média, desvio padrão, valor mínimo, valor máximo e quartis (primeiro, segundo e terceiro)

# **Produtividade e educação**

---

Análise e comparação dos dados estatísticos sobre os aplicativps de produtividade e os aplicativos de educação, para que possamos ter uma noção de como estão distribuídos na loja virtual.

Inicio da **Análise Exploratória de Dados**

calcula e exibe a distribuição percentual das categorias "Produtividade" e "Educação" no DataFrame df_educ_product, com base na coluna Category.

```python
contagens = df_educ_product['Category'].value_counts()
Produtividade = contagens.iloc[0]
Educação = contagens.iloc[1]

print(f'Total: {Educação + Produtividade} ')
print('Produtividade(%): {:.2f}'.format(Produtividade/ (Educação + Produtividade)*100))
print('Educação(%): {:.2f}'.format(Educação / (Educação + Produtividade)*100))
```

Cria um gráfico de barras para mostrar a quantidade de aplicativos em cada categoria (Category) do DataFrame df_educ_product, com rótulos bem formatados e o título do gráfico configurado.

```python
sns.countplot(x='Category', data=df_educ_product)
plt.title('Aplicativos de educação e produtividade', fontsize=17)
plt.ylabel('Quantidade', fontsize=13)
plt.xlabel('Categoria', fontsize=13)
plt.xticks(rotation=90)
plt.show()
```

O número de apps de produtividade é muito maior que os de educação, aproximadamente **75%** contra apenas **25%**.

Realizando uma análise de distribuição dos aplicativos classificados como pagos e grátis nas categorias de produtividade e educação no DataFrame df_educ_product.

```python
app_pago = df_educ_product[df_educ_product['Type'] == 'Paid']['Category'].value_counts()
produtividade_pago = app_pago.iloc[0]
educação_pago = app_pago.iloc[1]
app_gratis = df_educ_product[df_educ_product['Type'] == 'Free']['Category'].value_counts()
produtividade_gratis = app_gratis.iloc[0]
educação_gratis = app_gratis.iloc[1]
app_gratis_total = produtividade_gratis+educação_gratis
app_pago_total = produtividade_pago + educação_pago
```

**Aplicativos pagos**

```python
print(f'Produtividade: {produtividade_pago}')
print(f'Educação: {educação_pago}')
print('Total: {} ({:.2f})%'.format(app_pago_total, app_pago_total/ (app_pago_total + app_gratis_total)*100))
print('Produtividade(%): {:.2f}'.format(produtividade_pago / (produtividade_pago + educação_pago)*100))
print('Educação(%): {:.2f}'.format(educação_pago / (produtividade_pago + educação_pago)*100))
```

**Aplicativos gratuitos**

```python
print(f'Produtividade: {produtividade_gratis}')
print(f'Educação: {educação_gratis}')
print('Total: {} ({:.2f})%'.format(app_gratis_total, app_gratis_total / (app_gratis_total + app_pago_total)*100))
print('Produtividade(%): {:.2f}%'.format(produtividade_gratis/(produtividade_gratis + educação_gratis)*100))
print('Educação(%): {:.2f}%'.format(educação_gratis/(educação_gratis + produtividade_gratis)*100))
```

**Visualização dos dados**

```python
sns.countplot(x='Type', data=df_educ_product, hue='Category')
plt.title('Comparando aplicativos pagos x gratuitos entre educação e produtividade', fontsize=15)
plt.xlabel('Tipo', fontsize=13)
plt.ylabel('Quantidade', fontsize=13)
plt.show()
```

**Observação:** os aplicativos gratuitos de produtividade e educação correspondem a `95.35%` do total, enquanto `4.65%` são os aplicativos pagos.
Dos aplicativos pagos, temos `74.50%` correspondendo aos apps de produtividade, e `25.50%` aos apps de educação
